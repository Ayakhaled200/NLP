```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
```

```python
# Download NLTK resources (only required once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

###remove punvtuation, stopwords + make the text lower case

```python
# Define text preprocessing functions
def preprocess_text(text):
    # Tokenize text
    tokens = word_tokenize(text.lower())

    # Remove punctuation
    tokens = [token for token in tokens if token not in string.punctuation]

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join tokens back into a single string
    processed_text = ' '.join(tokens)

    return processed_text

# Apply text preprocessing to the 'context' column
df['Article Title'] = df['Article Title'].apply(preprocess_text)
```

```python
from sentence_transformers import SentenceTransformer

# Define SentenceTransformer model
model_st = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Define functions for embedding text
def get_sentence_transformers_embedding(text):
    embeddings = model_st.encode([text])
    return embeddings.flatten()

# Apply embedding techniques to the 'Article Title' column
df['sentence_transformers_embedding'] = df['Article Title'].apply(get_sentence_transformers_embedding)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Define X and y
X = df['sentence_transformers_embedding']  # Features
y = df['Category']  # Target variable

# Encode the target variable (y)
y_encoded = label_encoder.fit_transform(y)

# Split the encoded target variable into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train.tolist(), y_train)

# Predict on the testing data
y_pred_encoded = model.predict(X_test.tolist())

# Decode the predictions if needed
# y_pred = label_encoder.inverse_transform(y_pred_encoded)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred_encoded)
print("Mean Squared Error:", mse)
```

##Doc2Vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

# Tokenize article titles and create tagged documents
tagged_data = [TaggedDocument(words=title.split(), tags=[i]) for i, title in enumerate(df['Article Title'])]

# Train Doc2Vec model
doc2vec_model = Doc2Vec(vector_size=100, window=5, min_count=1, workers=4, epochs=20)
doc2vec_model.build_vocab(tagged_data)
doc2vec_model.train(tagged_data, total_examples=doc2vec_model.corpus_count, epochs=doc2vec_model.epochs)

# Function to generate Doc2Vec embeddings for each article title
def get_doc2vec_embedding(text):
    tokens = text.split()  # Assuming titles are already tokenized
    embedding = doc2vec_model.infer_vector(tokens)
    return embedding

# Apply Doc2Vec embedding to 'Article Title' column
df['doc2vec_embedding'] = df['Article Title'].apply(get_doc2vec_embedding)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Define X and y
X = np.stack(df['doc2vec_embedding'])  # Features
y = df['Category']  # Target variable

# Encode the target variable (y)
y_encoded = label_encoder.fit_transform(y)

# Split the encoded target variable into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_pred_encoded = model.predict(X_test)

# Decode the predictions
# y_pred = label_encoder.inverse_transform(y_pred_encoded)

# Calculate Mean Squared Error
y_pred= np.round(y_pred_encoded).astype(int)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

##Word2vec

```python
import gensim.downloader as api

# Load the pre-trained Word2Vec model
word2vec_model = api.load("word2vec-google-news-300")

def get_word2vec_embedding(text, model):
    tokens = text.split()  # Assuming titles are already tokenized
    embeddings = [model[token] for token in tokens if token in model]
    if embeddings:
        return sum(embeddings) / len(embeddings)
    else:
        return None

# Apply Word2Vec embedding to 'Article Title' column
df['word2vec_embedding'] = df['Article Title'].apply(lambda x: get_word2vec_embedding(x, word2vec_model))
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Define features (X) and labels (y)
X = np.asarray(df['word2vec_embedding'])

# Assuming X contains lists of word embeddings with variable lengths
max_length = 100  # Adjust as needed

# Pad or truncate the arrays in X to ensure they all have the same length
X_padded = pad_sequences(X, maxlen=max_length, padding='post', truncating='post')

# Convert X_padded to a NumPy array
X = np.array(X_padded)
y = np.asarray(df['Category'])
label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_pred = model.predict(X_test)

# Evaluate the model (optional)
# For regression tasks, common evaluation metrics include Mean Squared Error (MSE), R-squared, etc.
from sklearn.metrics import mean_squared_error, r2_score

y_pred= np.round(y_pred).astype(int)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

#FastText

```python
from gensim.models import FastText

# Train FastText model
fasttext_model = FastText(sentences=df['Article Title'], vector_size=100, window=5, min_count=1, workers=4)

# Function to generate FastText embeddings for each text
def get_fasttext_embedding(text):
    embeddings = [fasttext_model.wv[token] for token in text.split() if token in fasttext_model.wv]
    if embeddings:
        return sum(embeddings) / len(embeddings)
    else:
        return None

# Apply FastText embedding to the text column
df['fasttext_embedding'] = df['Article Title'].apply(get_fasttext_embedding)
```
